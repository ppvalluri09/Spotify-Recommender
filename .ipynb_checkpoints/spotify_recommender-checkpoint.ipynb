# Spotify Recommender

We are using the Million Song Dataset provided by spotify and cluster the data. Now we get the user's song data and find out the cluster the songs belong to annd recommend similar songs that belong to the same cluster.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/song_data.csv').iloc[:,1:]
df.head()
```

```python
features = df.copy().drop(['track_name', 'artist', 'key'], axis=1)
track_features = df[['track_name', 'artist']]
print(track_features.head())
features.head()
```

```python
fig = plt.figure(figsize=(16, 16))
color = ['r', 'g', 'b', 'c', 'm', 'y'] * 2
for i in range(features.shape[1]):
    plt.subplot(3, 3, i + 1)
    sns.distplot(features.iloc[:, i], color=color[i])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scl = StandardScaler().fit(features)
features = scl.transform(features)
features
```

```python
x_t, x_v = train_test_split(features, test_size=0.1, random_state=42)
print(x_t.shape, x_v.shape)
```

```python
from sklearn.cluster import KMeans
cost = []
for i in range(2, 8):
    kmeans = KMeans(n_clusters=2, max_iter=500)
    kmeans.fit(x_t)
    cost.append(kmeans.inertia_)
```

```python
plt.plot(range(1, len(cost) + 1), cost, c='g', label='Cost vs K value')
plt.xlabel('K Value')
plt.ylabel('Squared Error')
plt.show()
```

#### The optimal value of K seems to be 3, therefore we proceed with k = 3

```python
kmeans = KMeans(n_clusters=3, max_iter=500)
kmeans.fit(x_t)
```

```python
from sklearn.decomposition import PCA
decomposed = PCA(n_components=2).fit_transform(x_t)
decomposed
```

```python
centers = PCA(n_components=2).fit_transform(kmeans.cluster_centers_)
labels = kmeans.labels_
cluster1 = np.array([[decomposed[i, :]] for i in range(decomposed.shape[0]) if labels[i] == 0])
cluster2 = np.array([[decomposed[i, :]] for i in range(decomposed.shape[0]) if labels[i] == 1])
cluster3 = np.array([[decomposed[i, :]] for i in range(decomposed.shape[0]) if labels[i] == 2])
cluster3 = cluster3.reshape(cluster3.shape[0], cluster3.shape[2])
cluster2 = cluster2.reshape(cluster2.shape[0], cluster2.shape[2])
cluster1 = cluster1.reshape(cluster1.shape[0], cluster1.shape[2])
```

```python
plt.scatter(cluster1[:, 0], cluster1[:, 1], s=1, c='g', alpha=0.1, label='Cluster1')
plt.scatter(cluster2[:, 0], cluster2[:, 1], s=1, c='b', alpha=0.1, label='Cluster2')
plt.scatter(cluster3[:, 0], cluster3[:, 1], s=1, c='r', alpha=0.1, label='Cluster3')
plt.legend()
plt.show()
```

```python
import pickle
pickle.dump(kmeans, open('model.pickle', 'wb'))
```

```python
y_v = kmeans.predict(x_v)
y_v
```
